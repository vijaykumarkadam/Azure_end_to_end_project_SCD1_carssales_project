```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("parquet")\
        .load("abfss://bronze@storageforcarsproject.dfs.core.windows.net/rawData")
```

```python
df = df.withColumn("model_subcat",split(col("Model_ID"),"-")[0])
```

```python
df = df.withColumn("cost_per_unit",col("Revenue")/col("Units_Sold"))
```

```python
df = df.withColumn("Units_Sold",col("Units_Sold").cast(StringType()))
```

```python
df.write.format("parquet")\
    .mode("append")\
    .option("path","abfss://silver@storageforcarsproject.dfs.core.windows.net/OBT/")\
    .saveAsTable("cars_cata.silver.silverObtTable")
```
